```python

import pandas as pd

# Load dataset
df = pd.read_csv("../data/raw/CRM_Donor_Simulation_Dataset (2).csv")

# Preview data
df.head()
```

```python
# Check dataset structure
df.info()
```

## Data Inspection Summary

### What I ran

- `df.info()` — to inspect rows, columns, and data types
- `df.isnull().sum()` — to check missing values by column

### What I found

- The dataset has 5,000 rows and 13 columns
- No missing values were detected
- `LastDonationDate` was stored as text and needed conversion to datetime

### Action taken

I converted `LastDonationDate` to datetime so I could aggregate by month and year, and build date-based visuals in Power BI.

```python
# Convert LastDonationDate to datetime
df['LastDonationDate'] = pd.to_datetime(df['LastDonationDate'], errors='coerce')

# Check result
df.info()
```

### Data Cleaning Update

I converted `LastDonationDate` to a proper `datetime` type and verified the change with `df.info()`. This makes it possible to extract year and month for time-series analysis and to build accurate Power BI date visuals.

## Feature Engineering (What I Did)

I extracted `Year` and `Month` from `LastDonationDate` so the dataset can be quickly aggregated in Power BI and used for time-based filters.

### What I created

- `Year` — for year-over-year comparisons and KPI cards
- `Month` — for monthly trend charts and seasonal analysis

### Why I did it

Full date values made quick aggregation cumbersome. By adding `Year` and `Month` I could:

- Build the "Revenue Trend Over Time" chart in Power BI
- Apply dashboard filters at year or month level
- Speed up visual queries and simplify DAX measures

### Result

The cleaned dataset now includes `Year` and `Month`, which I use throughout the dashboard to control time-based views and filters.

```python
# Extract Year and Month from LastDonationDate
df['Year'] = df['LastDonationDate'].dt.year
df['Month'] = df['LastDonationDate'].dt.month

# Preview updated data
df.head()
```

## Customer Segmentation

I created a `Segment` column to classify donors by `TotalAmountDonated` so the dashboard can focus on priority groups.

### How I built it

- `High Value` — `TotalAmountDonated` > 15,000
- `Medium Value` — `TotalAmountDonated` between 7,000 and 15,000
- `Low Value` — `TotalAmountDonated` < 7,000

### Why I did it

Segmenting donors lets me build dashboard filters and visuals that highlight revenue drivers and targetable audiences (e.g., a Top Donors table or a segment-based KPI card).

```python
# Create customer segmentation
def segment(value):
    if value > 15000:
        return "High Value"
    elif value > 7000:
        return "Medium Value"
    else:
        return "Low Value"

df['Segment'] = df['TotalAmountDonated'].apply(segment)

# Preview result
df[['TotalAmountDonated', 'Segment']].head()
```

## Segmentation Result

### What I did

I added the `Segment` column and checked that donors fall into High, Medium, and Low categories according to donation totals.

### What I observed

- High-value donors account for the largest share of revenue
- Medium-value donors provide meaningful contribution and growth opportunity
- Low-value donors contribute less but are useful for broad engagement strategies

### How I use this in Power BI

I use the `Segment` column as a slicer and to build charts that compare revenue and counts by segment, helping stakeholders focus on retention and acquisition strategies.

## Revenue Analysis

I calculated total donation revenue to provide a baseline KPI for the dashboard and to measure overall performance before breaking results down by segment and time.

```python
# Total Revenue
total_revenue = df['TotalAmountDonated'].sum()


print(f"Total Revenue: ${total_revenue:,.2f}")
```

## Revenue Insight

### What I did

I summed `TotalAmountDonated` to compute total revenue and used that value as a KPI in the Power BI dashboard.

### Why this matters

Total revenue provides a top-level performance metric and a baseline for comparing segments, months, and donor cohorts.

## Top Donors Analysis

I identified the top 10 donors by sorting `TotalAmountDonated` in descending order. This helps populate a Top Donors table in the dashboard and informs donor outreach priorities.

```python
# Top 10 donors by total donation
top_donors = df.sort_values(by='TotalAmountDonated', ascending=False).head(10)

# Show result
top_donors[['FirstName', 'LastName', 'TotalAmountDonated']]
```

## Top Donors Insight

### What I found

The top donors contribute a disproportionately large share of revenue, with several individual donations near the dataset's maximum values.

### How I use this

I surface the top donors on the dashboard and use the list to recommend targeted retention and engagement actions.

## Engagement Analysis

I analyzed average donation by `EngagementScore` to check whether more engaged donors give more. This informs which engagement efforts are most likely to improve revenue.

### What I did

- Grouped donors by `EngagementScore`
- Calculated average `TotalAmountDonated` per engagement level

### Why this matters for the dashboard

These results feed a visual that links engagement to donation behavior, helping the team prioritize engagement programs.

```python
# Average donation by engagement score
engagement_analysis = df.groupby('EngagementScore')['TotalAmountDonated'].mean()

# Show first few results
engagement_analysis.head()
```

## Engagement Insight

### What I observed

Donors with higher `EngagementScore` tend to donate more on average.

### How I use this

I use the engagement analysis in the dashboard to identify high-potential donors and to justify engagement-focused initiatives that can improve revenue.

```python
# Revenue by segment
segment_revenue = df.groupby('Segment')['TotalAmountDonated'].sum()

# Show result
segment_revenue
```
