## Feature Engineering

###  What is Feature Engineering?

Feature engineering is the process of creating new columns (features) from existing data to make analysis easier and more meaningful.



###  Why is Feature Engineering Needed?

- Raw data is often not ready for analysis directly  
- It helps transform data into a format suitable for insights and visualization  
- It enables deeper analysis such as trends, patterns, and segmentation  



###  What We Are Creating

From the column `LastDonationDate`, we will extract:

- **Year** → to analyze yearly donation trends  
- **Month** → to analyze monthly patterns  


###  Why These Features Are Important

- Allows analysis of **time-based trends** (e.g., yearly growth, seasonal patterns)  
- Helps build **Power BI dashboards with time filters**  
- Enables grouping data by **month and year**  
- Supports decision-making by showing changes over time  


###  Expected Outcome

The dataset will include new columns:
- `Year`
- `Month`

These features will be used in further analysis and dashboard creation.


```python
# Extract Year and Month from LastDonationDate
df['Year'] = df['LastDonationDate'].dt.year
df['Month'] = df['LastDonationDate'].dt.month

# Preview updated data
df.head()
```

## Customer Segmentation

### What is Customer Segmentation?

Customer segmentation means grouping customers based on their value or behavior.



### Why is Segmentation Important?

- Helps identify high-value customers
- Supports better business decisions
- Improves marketing and engagement strategies
- Makes analysis more meaningful



### What We Will Do

We will create a new column called `Segment` based on the total donation amount.

Customers will be grouped into:

- High Value → donation greater than 15000
- Medium Value → donation between 7000 and 15000
- Low Value → donation less than 7000



### Why This Is Needed

- Helps understand which customers contribute the most revenue
- Useful for Power BI dashboards and filtering
- Supports customer targeting and strategy



### Expected Result

A new column called `Segment` will be added to the dataset with values:
- High Value
- Medium Value
- Low Value



```python
# Create customer segmentation
def segment(value):
    if value > 15000:
        return "High Value"
    elif value > 7000:
        return "Medium Value"
    else:
        return "Low Value"

df['Segment'] = df['TotalAmountDonated'].apply(segment)

# Preview result
df[['TotalAmountDonated', 'Segment']].head()
```

## Segmentation Result

### What Was Done

A new column called `Segment` was created to classify customers based on their total donation amount.


### What We See

- Customers are now grouped into:
  - High Value
  - Medium Value
  - Low Value

- The segmentation correctly reflects donation levels:
  - Higher donation → High Value
  - Lower donation → Low Value



### Why This Matters

- Helps identify the most important customers
- Makes it easier to analyze revenue contribution
- Useful for building dashboard filters and insights
- Supports business decisions such as targeting high-value customers

---

### Conclusion

The dataset is now enhanced with meaningful categories and is ready for deeper analysis and visualization.

## Engagement Analysis

### What We Are Doing

We analyze how customer engagement relates to donation behavior.



### Why This Is Important

- Helps understand customer activity and involvement

```python
# Average donation by engagement score
engagement_analysis = df.groupby('EngagementScore')['TotalAmountDonated'].mean()

# Show first few results
engagement_analysis.head()
```
